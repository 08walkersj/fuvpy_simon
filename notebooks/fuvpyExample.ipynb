# How to use fuvpy

fuvpy is a collection of functions to analyse and visualize auroral images.
The images are stored as xarray.Datasets and fuvpy includes a function to read .idl/.sav file generated with fuview3 or xvis (IMAGE and Polar softwave) into this format.

Images are loaded as follows:

```python
import fuvpy as fuv
import pandas as pd
import glob
import warnings
import matplotlib.pyplot as plt
from polplot import pp


path = '../data/wicFiles/'
wicfiles = glob.glob(path + '*.idl')
wic = fuv.readImg(wicfiles)
wic
```

There are two ways to select images:

```python
wic.sel(date='2000-08-28 09:28:42')['img'].plot()
```

```python
wic = fuv.getRayleigh(wic,'img')
wic.isel(date=[0,2,4,7])['imgR'].plot(x='col', y='row', col='date', col_wrap=2,robust=True)
```

In the same way we can visualize for instance the viewing anlge for these images:

```python
wic.isel(date=slice(4,8))['dza'].plot(x='col', y='row', col='date', col_wrap=2)
```

Note how the name and units of the variables appear automatically in the figures. They are added to each variable, and can be accessed (and assigned) using wic['nameofvariable'].attrs.

A routine to plot the images in polar coordinates has been added to the Polarsubplot class.
Here is an example of how to use it:

```python
fig = plt.figure(figsize=(4,5))
ax = plt.subplot2grid((10, 1), (0, 0),rowspan=9)
pax = pp(ax)
fuv.showImg(wic.isel(date=4),'imgR',pax=pax,crange=(0,9),cmap='plasma')

cax = plt.subplot2grid((10, 10), (9, 1),rowspan=1,colspan=8)
cax.axis('off')
cbar = plt.colorbar(pax.ax.collections[0],orientation='horizontal',ax=cax,fraction=1,extend='max')
cbar.set_label('{} ({})'.format(wic['imgR'].attrs['long_name'],wic['imgR'].attrs['units']))

ax.set_title(wic['id'].values.tolist() + ': ' + 
             wic.isel(date=2)['date'].dt.strftime('%Y-%m-%d %H:%M:%S').values.tolist())
# It should really be possible to simplify this line
plt.tight_layout()
```

Time to remove the dayglow. A simple version of the method is described in the appendix in *Ohma et al.* (2018), but the present version is more advanced and inludes a temporal fit using B-slines. Going retro, we use the colormap from the IMAGE mission. 

```python
wic = fuv.makeDGmodel(wic,transform='log',tOrder=0)
fuv.showDGmodel(wic.isel(date=6),cmap=fuv.getIMAGEcmap())
```

Dayglow is subtracted, but the image is not perfect (note the asymmetry at the dayside. We model the residual background using a spheric harmonics model. Note the use of Dataset.pipe, which is handy for chained function calls.

```python
wic = wic.pipe(fuv.makeSHmodel,4,4,order=0)

fig = plt.figure(figsize=(4,5))
ax = plt.subplot2grid((10, 1), (0, 0),rowspan=9)
pax = pp(ax)
fuv.showImg(wic.isel(date=6),'shimg',pax=pax,crange=(-1000,1000),cmap='seismic')

cax = plt.subplot2grid((10, 10), (9, 1),rowspan=1,colspan=8)
cax.axis('off')
cbar = plt.colorbar(pax.ax.collections[0],orientation='horizontal',ax=cax,fraction=1,extend='max')
cbar.set_label('{}'.format(wic['shimg'].attrs['long_name']))

ax.set_title(wic['id'].values.tolist() + ': ' + 
             wic.isel(date=2)['date'].dt.strftime('%Y-%m-%d %H:%M:%S').values.tolist())
# It should really be possible to simplify this line
plt.tight_layout()
```

To further examine the performance of the background subtraction, we plot the background corrected image (residual counts) vs solar zenith angle and modelled image:

```python
plt.figure()
wic.plot.scatter(x='sza',y='shimg',hue='shweight',s=0.01,alpha=0.4)
wic['totmodel'] = wic['dgmodel'] + wic['shmodel']
plt.figure()
wic.plot.scatter(x='totmodel',y='img',hue='shweight',s=0.01,alpha=0.4)
```

The images can also be plotted in an mlt-mlat projection. If there is any interest, this function can be expanded to also do mlon-mlat and glon-glat projections.

```python
fuv.showImgProj(wic.isel(date=4),inImg='shimg',crange=(0,1500),cmap=fuv.getIMAGEcmap())
```

A two step approach to model the auroral boundaries is included in fuvpy. The first step identifies significant peaks in 24 1-wide MLT sectors in each image individually. Based on these boundaries, the second step makes a spatiotemporal model of the boundaries using radial Fourier serier (space) and B-splines (time). This model is smooth and differentiable at any MLT location in the time interval spanned by the input images.

```python
import numpy as np
b = fuv.findBoundaries(wic)
fuv.ppBoundaries(b.sel(lim=1),'ocb')
```

```python
b2 = fuv.makeBoundaryModel(b)
fuv.ppBoundaries(b2)
```

```python
flux= fuv.calcFlux(b2)
flux['openFlux'].plot()
```
